```
class Libro:
    def __init__(self, titulo: str, autor: str, isbn: str, precio: float):
        self._titulo = titulo
        self._autor = autor
        self._isbn = isbn
        self._precio = precio

    @property
    def titulo(self) -> str:
        """Obtiene el titulo del libro."""
        return self._titulo

    @titulo.setter
    def titulo(self, valor: str) -> None:
        """Establece el titulo del libro con validacion"""
        if isinstance(valor, str) and valor.strip():
            self._titulo = valor
        else:
            raise ValueError("EL TITULO DEBE SER UN TEXTO VALIDO NO VACIO")

    @property
    def autor(self) -> str:
        """Obtiene el autor del libro."""
        return self._autor

    @autor.setter
    def autor(self, valor: str) -> None:
        """Establece el autor del libro con validacion"""
        if isinstance(valor, str) and valor.strip():
            self._autor = valor
        else:
            raise ValueError("EL AUTOR DEBE SER UN TEXTO VALIDO NO VACIO")

    @property
    def isbn(self) -> str:
        return self._isbn

    @isbn.setter
    def isbn(self, valor: str) -> None:
        """Establece el ISBN del libro con validacion"""
        if isinstance(valor, str) and valor.strip():
            self._isbn = valor
        else:
            raise ValueError("EL ISBN DEBE SER UN TEXTO VALIDO NO VACIO")


    @property
    def precio(self) -> float:
        return self._precio

    @precio.setter
    def precio(self, valor: float) -> None:
        """Establece el precio del libro con validacion"""
        if isinstance(valor, (int, float)) and valor > 0:
            self._precio = float(valor)
        else:
            raise ValueError("EL PRECIO DEBE SER UN NUMEOR POSITIVO.")




    def __repr__(self) -> str:
        #Representacion segun estaba oop y pep#
        return (
            f"Libro(titulo='{self._titulo}', autor='{self._autor}', "
            f"isbn='{self._isbn}', precio={self._precio})"
        )


# MAIN (DEMOSTRACION)

def main():
  print("\n=== Informacion del libro ===")
  libro1 = Libro("python IA","pedirot perez","123-456-789",120000.00)
  print(libro1)

  print("\n=== Dtos Actualizados ===")

  libro1.titulo = "Python Avanzado"
  libro1.precio = 150000.00
  print(libro1)

if __name__ == "__main__":
  main()
```

```
class Opereacion():
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2

  def calcular(self):
      print("*******ALERTA A INGRESAR OPERACIONES ATENTOS,*******")

class suma(Opereacion):
  def __init__(self, num1, num2):
    super().__init__(num1, num2)
    print("preparando la opereacio suma")
  def calcular(self):
    super().calcular()
    resultado = self.num1 + self.num2
    print(f"El resultado de la suma es: {resultado}")

# OPEREACION RESTA
class Resta(Opereacion):
  def __init__(self, num1, num2):
    super().__init__(num1, num2)
    print("preparando la opereacio resta")
  def calcular(self):
    super().calcular()
    resultado = self.num1 - self.num2
    print(f"El resultado de la resta es: {resultado}")
# OPEREACION MULTIPLICACION

class Multiplcacion(Opereacion):
  def __init__(self, num1, num2):
    super().__init__(num1, num2)
    print("preparando la opereacio multiplicacion")
  def calcular(self):
    super().calcular()
    resultado = self.num1 * self.num2
    print(f"El resultado de la multiplicacion es: {resultado}")

# OPEREACION DIVICION
class Divicion(Opereacion):
  def __init__(self, num1, num2):
    super().__init__(num1, num2)
    print("preparando la opereacio divicion")
  def calcular(self):
    super().calcular()
    if self.num2 != 0:
        resultado = self.num1 / self.num2
        print(f"El resultado de la divicion es: {resultado}")
    else:
        print("Error: No se puede dividir por cero.")

def main():
    print("\n--- Operación Suma ---")
    operacion_suma = suma(20,30)
    operacion_suma.calcular()
    print("ya paso la alerta fin de la operacion")

    print("\n--- Operación Resta ---")
    operacion_resta = resta(20,120)
    operacion_resta.calcular()
    print("ya paso la alerta fin de la operacion")

    print("\n--- Operación Multiplicación ---")
    operacion_multiplicacion = Multiplcacion(90,30)
    operacion_multiplicacion.calcular()
    print("ya paso la alerta fin de la operacion")

    print("\n--- Operación División ---")
    operacion_division = Divicion(20,99)
    operacion_division.calcular()
    print("ya paso la alerta fin de la operacion")

    print("\n--- Operación División por Cero ---")
    operacion_division_cero = Divicion(10,0)
    operacion_division_cero.calcular()
    print("ya paso la alerta fin de la operacion")


if __name__ == "__main__":
  main()
```

```
from abc import ABC, abstractmethod

class ProcesoDePago(ABC):
  @abstractmethod
  def procesar_pago(self, monto: float):
    pass

class PagoEfectivo(ProcesoDePago):
  def __init__(self, monto: float):
    self.monto = monto

  def procesar_pago(self, monto: float):
    print(f"Realizando proceso de  pago en efectivo por: ${monto:,.2f}")

class PagoTarjeta(ProcesoDePago):
  def __init__(self, monto: float):
    self.monto = monto

  def procesar_pago(self, monto: float):
    print(f"Realizando el Reporte  de  pago con tarjeta por: ${monto:,.2f}")


def realizar_pago(metodo_pago: ProcesoDePago, monto: float):
  metodo_pago.procesar_pago(monto)

def main():
  print("\n--- Realizando el  Proceso de Pagos ---\n")

  pago_efectivo_obj = PagoEfectivo(1000)
  realizar_pago(pago_efectivo_obj, 1000)

  pago_tarjeta_obj = PagoTarjeta(2000)
  realizar_pago(pago_tarjeta_obj, 2000)


  realizar_pago(pago_efectivo_obj, 550.75)

if __name__ == "__main__":
  main()
```

```
from abc import ABC, abstractmethod

class Figura(ABC):
    @abstractmethod
    def calcular_area(self):
        pass

class Circulo(Figura):
  def __init__(self, radio):
    self.radio = radio
  def calcular_area(self):
    return (3.14 *( self.radio ** 2))

class Rectangulo(Figura):

  def __init__(self, ancho, alto):
    self.ancho = ancho
    self.alto = alto
  def calcular_area(self):
    return self.ancho * self.alto

class Cuadrado(Figura):
  def __init__(self, lado):
    self.lado = lado
  def calcular_area(self):
    return self.lado * self.lado

class Triangulo(Figura):

  def __init__(self, base, altura):
    self.base = base
    self.altura = altura
  def calcular_area(self):
    return 0.5 * self.base * self.altura

class Cubo(Figura):

  def __init__(self, lado):
    self.lado = lado

  def calcular_area(self):

    return 6 * (self.lado ** 2)

class Piramide(Figura):

  def __init__(self, base_lado, altura_inclinada=None):
    self.base_lado = base_lado
    self.altura_inclinada = altura_inclinada
  def calcular_area(self):

    return self.base_lado ** 2

def mostrar_area(figura: Figura):
  print(f"El area: {figura.calcular_area():.2f}")

def main():
  circulo1 = Circulo(5)
  mostrar_area(circulo1)

  rectangulo1 = Rectangulo(3, 7)
  mostrar_area(rectangulo1)

  cuadrado1 = Cuadrado(4)
  mostrar_area(cuadrado1)

  triangulo1 = Triangulo(3, 5)
  mostrar_area(triangulo1)

  cubo1 = Cubo(4)
  mostrar_area(cubo1)

  piramide1 = Piramide(6)
  mostrar_area(piramide1)

if __name__ == "__main__":
  main()
```

```
from abc import ABC, abstractmethod

class Empleado(ABC):
  def __init__(self, nombre: str) -> None:
    self.nombre = nombre

  @property
  def nombre(self) -> str:
    return self._nombre

  @nombre.setter
  def nombre(self, valor: str) -> None:
    if isinstance(valor, str) and valor.strip() != "":
      self._nombre = valor.strip()
    else:
     raise ValueError("El nombre debe ser un texto valido")

  @abstractmethod
  def trabajar(self) -> None:
      pass

class Gerente(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} esta realizando la administracion de la empresa")

class Desarrollador(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} esta programando una solucion en python")

def ejecutar_tarea(empleado: Empleado) -> None:
  empleado.trabajar()

class vendedor(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} esta vendiendo productos de la empresa")


class tecnico(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} esta arreglando productos de la empresa")

class secretario(Empleado):
  def trabajar(self) -> None:

def main():
  print("***********INICIO DEL PROGRAMA DE LA EMPLEADO SENA************")

  gerente1 = Gerente("santiago")
  ejecutar_tarea(gerente1)

  desarrollador1 = Desarrollador("Kevin")
  ejecutar_tarea(desarrollador1)

  print("\n-------Probando validación de nombre-------")
  try:

    desarrollador_invalido = Desarrollador("SANTIAGO")
    ejecutar_tarea(desarrollador_invalido)
  except ValueError as e:
    print(f"Error esperado: {e}")

  try:
    gerente_invalido = Gerente(123)
    ejecutar_tarea(gerente_invalido)
  except ValueError as e:
    print(f"Error inesperado: {e}")


if __name__ == "__main__":
  main()
```

```
from abc import ABC, abstractmethod

class Generar_Reporte(ABC):
  @abstractmethod
  def generar_reporte(self)-> None:
    pass

class Sistema_de_pago(ABC):
  def __init__(self, monto: float) -> None:
    self.monto = monto

  @abstractmethod
  def procesar_pago(self)-> None:
    pass

class PagoBancario(Sistema_de_pago, Generar_Reporte):
  def procesar_pago(self)-> None:
    print(f"procesando transferencia bancaria por ${self.monto:.2f}")

  def generar_reporte(self)-> None:
    print(f"reporte de pago realizado mediante el sistema bancario")

class PagoCripto(Sistema_de_pago):
  def procesar_pago(self)-> None:
    print(f"procesar pago cripto moneda por ${self.monto:.2f}")

  def generar_reporte(self)-> None:
    print(f"reporte de pago realizado mediante criptomoneda")

def ejecutar_pago(pago: Sistema_de_pago) -> None:
  pago.procesar_pago()

def main() -> None:
  print("\n--- Realizando el Proceso de Pagos ---\n")

  pago_bancario1 = PagoBancario(1000)
  ejecutar_pago(pago_bancario1)
  pago_bancario1.generar_reporte()

  pago_cripto1 = PagoCripto(2000)
  ejecutar_pago(pago_cripto1)
  pago_cripto1.generar_reporte()

if __name__ == "__main__":
  main()
```

```
from abc import ABC, abstractmethod
from typing import Optional

class Movibles(ABC):
  @abstractmethod
  def mover(self)-> None:
    pass

class Animal(ABC):
  def __init__(self, nombre: str) -> None:
     self.__nombre: str = ""
     self.nombre = nombre
  @property
  def nombre(self) -> str:
    return self.__nombre

  @nombre.setter
  def nombre(self, valor: str) -> None:

    if isinstance(valor, str) and valor.strip():
      self.__nombre = valor.strip().title()

    else:
      raise ValueError("El nombre debe ser un texto valido")


  @abstractmethod
  def sonido(self) -> None:
    pass


class Perro(Animal):
  def sonido(self)-> None:
      print(f"{self.nombre} (perro)dice:Guuu Guuauu ")

class Gato(Animal):
  def sonido(self)-> None:
      print(f"{self.nombre} (gato)dice:Miauuu")

class Vaca(Animal):
  def sonido(self)-> None:
      print(f"{self.nombre} (vaca)dice:Muuuu")


class Caballo(Animal):
  def sonido(self)-> None:
      print(f"{self.nombre} (caballo)dice:Hooooooooo")

class Leon(Animal, Movibles):
  def sonido(self)-> None:
      print(f"{self.nombre} (leon)dice:ROARRRRRR")

  def mover(self)-> None:
    print(f"{self.nombre} camina Sigilosamente por la sabana,,")

def hacer_sonido(animal: Animal) -> None:
    animal.sonido()

def desplazar (animal: Movibles) -> None:
    animal.mover()

def main() -> None:
    try:

      animales =[
          Perro("Rocky"),
          Gato("Luna"),
          Vaca("Lola"),
          Caballo("Thunder"),
          Leon("Simba")
      ]
      print("==POLIMORFISMO EN MODELO ZOOLOGICO ===\n")
      for animal in animales:
        hacer_sonido(animal)
      print("\n MOVIMIENTO +++")
      for animal in animales:
        if isinstance(animal, Movibles):
          desplazar(animal)

      print(f"\nActualizacion de los nombres de los animales")
      for animal in animales:
        animal.nombre = "MAX"

      print(f"el nuevo nombre del primer animal es {animales[0].nombre}")

    except ValueError as e:

      print(f"Error: {e}")


if __name__ == "__main__":
    main()
```

```
from abc import ABC, abstractmethod

# Clase abstracta
class Vendedor(ABC):

    def __init__(self, nombre, apellidos):
        self.nombre = nombre
        self.apellidos = apellidos

    @abstractmethod
    def vender(self):
        pass

    def mover(self):
        print(f"{self.nombre} se está desplazando para vender")


class VendedorTienda(Vendedor):

    def vender(self):
        print(f"{self.nombre} vende en la tienda física")


class VendedorOnline(Vendedor):

    def vender(self):
        print(f"{self.nombre} vende por internet")


class VendedorPromotor(Vendedor):

    def vender(self):
        print(f"{self.nombre} promociona productos y vende")


class VendedorDomicilio(Vendedor):

    def vender(self):
        print(f"{self.nombre} vende en domicilio")

    def mover(self):
        print(f"{self.nombre} se mueve a la casa del cliente")


def main():

    v1 = VendedorTienda("Carlos", "Lopez")
    v2 = VendedorOnline("Ana", "Martinez")
    v3 = VendedorPromotor("Luis", "Ramirez")
    v4 = VendedorDomicilio("Sofia", "Gomez")

    vendedores = [v1, v2, v3, v4]

    for v in vendedores:
        v.vender()
        v.mover()


if __name__ == "__main__":
    main()
```
